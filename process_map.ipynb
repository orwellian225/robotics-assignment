```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('maps/map2.pgm', 0)
```

```python
plt.imshow(image, cmap='gray')
plt.show()
```

```python
t, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Define the structuring element
kernel = np.ones((10, 10), np.uint8)

# Perform dilation
eroded_image = cv2.erode(thresh, kernel, iterations=1)
```

```python
plt.imshow(eroded_image, cmap='gray')
plt.axis('off')
plt.show()  
```

```python
new_image=np.zeros(eroded_image.shape)
for i in range(eroded_image.shape[0]):
    for j in range(eroded_image.shape[1]):
        if (eroded_image[i][j]==False):
            # 1 is obstacles
            new_image[i][j]=1
        else:
            # 0 is open space
            new_image[i][j]=0


# save new image as png
print(new_image.shape)
# export edited file
cv2.imwrite("newimg.png", new_image*255)

#  save image array to text file
np.savetxt("mapArr.txt", new_image, delimiter =",",fmt='%d')
```
